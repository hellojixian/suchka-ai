```python
import pickle
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)

from sklearn.metrics.pairwise import cosine_similarity
data_folder = '/home/jixian/github/suchka-ai/output/faces/Jessy Jones/temp/11877003'
data_folder = '/home/jixian/github/suchka-ai/output/faces/Jessy Jones/temp/84624209'
data_folder = '/home/jixian/github/suchka-ai/output/faces/Jessy Jones/temp/39603925'

data_file = f'{data_folder}/embeddings.pkl'
with open(data_file, 'rb') as f:
  data = pickle.loads(f.read())

emotions_file = f'{data_folder}/emotions.pkl'
with open(emotions_file, 'rb') as f:
  emotions = pickle.loads(f.read())  
```

```python
m = cosine_similarity(list(data.values()))
df = pd.DataFrame(m)
n = np.array(m)
# drop values that equal to 1
n = n[n < 0.9999]
print(df)
print(np.min(n), np.mean(n), np.max(n))
```

```python
def cluster_embeddings(embeddings, threshold):
  groups = []
  data = embeddings
  keys = list(data.keys())
  visited = set()
  m = cosine_similarity(list(data.values()))
  for i in range(len(m)):
    if i in visited: continue
    # print(i, keys[i].replace(' ','\ '))
    group = dict()
    group[keys[i]] = data[keys[i]]
    for j in range(len(m)):
      if j in visited: continue
      if i == j: continue    
      similarity = m[i][j]
      if similarity >= threshold:
        visited.add(i)
        visited.add(j)
        group[keys[j]] = data[keys[j]]
        # print(j, similarity,keys[j].replace(' ','\ '))
    if len(group) > 1: groups.append(group)
  return groups

threshold = 0.65
groups = cluster_embeddings(data, threshold)

import shutil, os

i = 0
for group in groups:
  dst_path = f'{data_folder}/face_{i}'
  if not os.path.exists(dst_path): os.mkdir(dst_path)
  for f in group.keys():
    print(f.replace(' ','\ '))        
    shutil.copy(f, dst_path)
  print('---')
  i += 1
```
